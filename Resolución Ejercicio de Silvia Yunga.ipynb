```python
import numpy as np
import math

k=10 # numero de iteraciones
m,n = 1000, 1000 # Tamao de la matriz
L = -0.3
K = 5
A = np.eye(m,n, k= -1) * (-L) + np.eye(m,n) * K + np.eye(m, n, k=1) * (-L)
print(A)
#check that all eigenvalues are positive:

b = np.ones((n), dtype=float)
b[0] = 2
b[n-1] = 2
b[1:n-1] = 4
#print(b)
n=A.shape[1] #numero de columnas de los coeficientes de la matriz A
D=np.eye(n)
D[np.arange(n),np.arange(n)]=A[np.arange(n),np.arange(n)]
LU=D-A
L=np.tril(LU)
U=np.triu(LU)
#T=np.dot(np.linalg.inv(D),LU)
T=np.dot(np.linalg.inv(D),LU)
v=(np.linalg.eigvals(T))
rho_v=max(abs(v))
w=2/(1+math.sqrt(1-rho_v**2))
D_wL=D-w*L
x=np.zeros(n)
if rho_v <1:
    print(" ",0,"El resultado de la segunda iteracion:",x)
    for i in range(k):
        D_wL_inv=np.linalg.inv(D_wL)
        x=np.dot(np.dot(D_wL_inv,(1-w)*D+w*U),x)+w*np.dot(D_wL_inv,b)
        print(" ", i+1,"El resultado de la iteracion:",x.round(decimals=2))
else:
    print('Matrix no converge')
            
print(w)
```

#### La condicin necesaria y suficiente para que el proceso iterativo converja es que la matriz B converja a 0, y la condicin necesaria y suficiente para que la matriz B converja a 0 es que el radio espectral de la matriz B sea menor que 1.
